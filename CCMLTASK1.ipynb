* Spread and Distribution:
The interquartile ranges (the boxes) and overall spread (whiskers) are also similar between the two groups for each grade. Both groups show a wide range of scores, indicating considerable individual variation in academic performance regardless of relationship status.

* Outliers:
Both groups have a few outliers, including some students with very low grades. Outliers appear in both relationship categories, suggesting that extreme academic outcomes are not exclusive to either group.

* Consistency Across Terms:
The pattern holds consistently for G1, G2, and G3, indicating that the relationship status does not become more or less associated with academic performance as the year progresses



There is no significant difference in academic performance (G1, G2, G3) between students who are in romantic relationships and those who are not. Both groups display similar medians, ranges, and outlier patterns for all three grading periods. This indicates that, within this dataset, being in a romantic relationship does not appear to have a measurable impact—positive or negative—on students’ academic outcomes  (Only the ones in romantic relationship have slightly lesser grades)

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
from scipy.stats import randint
from sklearn.model_selection import RandomizedSearchCV
from sklearn import model_selection
features2 = numeric_df.drop(columns=['romantic']).columns
X = numeric_df[features2]
y = numeric_df['romantic']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

FEATURE SCALING

```python
%%capture
from sklearn.preprocessing import StandardScaler
sta = StandardScaler()

X_train.iloc[:, :] = sta.fit_transform(X_train.iloc[:, :])
X_test.iloc[:, :] = sta.transform(X_test.iloc[:, :])

print("done")
```

```python
#SOURCE :https://www.datacamp.com/tutorial/understanding-logistic-regression-python
#LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=10)
logreg.fit(X_train, y_train)
y_pred_log = logreg.predict(X_test)
print(confusion_matrix(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))
```

```python
#SOURCE : https://www.analyticsvidhya.com/blog/2021/06/5-hyperparameter-optimization-techniques-you-must-know-for-data-science-hackathons/
#Random Forest

from xgboost import XGBClassifier
clf = XGBClassifier()

```

```python
#GRIDSEARCHCV
%%capture
from sklearn import preprocessing
from sklearn import model_selection
from sklearn import metrics

param_grid = {
    "n_estimators":[100,200,300,400],
    "max_depth":[1,3,5,7],
    "reg_lambda":[.01,.1,.5]
}

model = model_selection.GridSearchCV(
    estimator = clf,
    param_grid = param_grid,
    scoring = 'accuracy',
    verbose = 10,
    n_jobs = 1,
    cv = 5
)
model.fit(X_train,y_train)

```

```python

y_pred_gcv = model.predict(X_test)
print(confusion_matrix(y_test, y_pred_gcv))
print(classification_report(y_test, y_pred_gcv))
```

```python
#RandomizedSearchCV
%%capture
param_grid = {
    "n_estimators":[100,200,300,400],
    "max_depth":[1,3,5,7],
    "reg_lambda":[.01,.1,.5]
}


model = model_selection.RandomizedSearchCV(
    estimator = clf,
    param_distributions = param_grid,
    scoring = 'accuracy',
    verbose = 10,
    n_jobs = 1,
    cv = 5,
    n_iter=10
)
model.fit(X_train,y_train)
```

```python
y_pred_rcv = model.predict(X_test)
print(confusion_matrix(y_test, y_pred_rcv))
print(classification_report(y_test, y_pred_rcv))
```

```python
#Naive Bayes Classification
#SOURCE : https://www.kaggle.com/code/prashant111/naive-bayes-classifier-in-python
# train a Gaussian Naive Bayes classifier on the training set
from sklearn.naive_bayes import GaussianNB


# instantiate the model
gnb = GaussianNB()


# fit the model
gnb.fit(X_train, y_train)
```

```python
y_pred_gnb = gnb.predict(X_test)

y_pred_gnb
```

```python
print(confusion_matrix(y_test, y_pred_gnb))
print(classification_report(y_test, y_pred_gnb))
```

```python
X = numeric_df[['G3', 'Dalc']]
y = numeric_df['romantic']

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

def plot_boundary(clf, name):
    clf.fit(X, y)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(6, 5))
    plt.contourf(xx, yy, Z, cmap='cool', alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolors='m')
    plt.xlabel('G3')
    plt.ylabel('Dalc')
    plt.title(name)
    plt.tight_layout()
    plt.show()



plot_boundary(LogisticRegression(), "Logistic Regression")
plot_boundary(GaussianNB(), "Gaussian Naive Bayes")
plot_boundary(KNeighborsClassifier(n_neighbors=5), "KNN")
plot_boundary(RandomForestClassifier(n_estimators=100), "Random Forest")
```

# Logistic Regression  

- The decision boundary is a straight line, reflecting the linear nature of logistic regression. This means the model separates classes based on a linear combination of the features (here, G3 and Dale).  
- In SHAP terms, feature contributions are additive and straightforward: each feature's effect on the prediction is proportional to its value and the learned weight: \( \overline{y} \) in \( \overline{u} \).  
- The magenta and cyan regions show the two predicted classes; points near the boundary are more sensitive to small changes in feature values, which is reflected in higher SHAP values for those features in those instances.  

# Gaussian Naive Bayes  

- The decision boundary is curved and more circular, reflecting the probabilistic and distributional assumptions of Naive Bayes (features are assumed independent and normally distributed within each class) \( \overline{w} \).  
- SHAP values here reflect the influence of each feature under the Gaussian assumption. The model's prediction for a point is determined by how likely its G3 and Dale values are under each class's Gaussian distribution.  
- The large cyan region in the lower right shows where the model is most confident for one class, while the magenta region dominates elsewhere. SHAP values in Naive Bayes can sometimes be less informative if the model's assumptions don't fit the data well: \( \overline{u} \).  

# KNN (K-Nearest Neighbors)  

- The decision boundary is highly irregular and follows the distribution of the training points closely, reflecting KNN's non-parametric, instance-based nature.  
- SHAP values for KNN are more complex to interpret, as predictions depend on the local neighborhood of each point. Feature contributions can be very local and change abruptly as you move in feature space \( \overline{u} \).  
- The patchwork pattern shows that small changes in G3 or Dale can flip the prediction, leading to potentially high SHAP values in regions near the boundary.  

# Random Forest  

- The decision boundary is blocky and piecewise, reflecting the ensemble of decision trees that make up the random forest. Each tree splits the feature space along axis-aligned lines, and the ensemble averages these splits \( \overline{w} \): \( \overline{u} \).  
- SHAP values for random forests (computed with TreeSHAP) can capture both main effects and feature interactions, providing insight into which features drive predictions in each region of the space.  
- The blocky cyan and magenta regions indicate areas where the majority of trees agree on the class, with boundaries determined by the splits in the trees.
